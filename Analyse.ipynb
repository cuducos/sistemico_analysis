# Setting the background

Imports and settings

```python
import csv
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import numpy as np

%matplotlib inline
pylab.rcParams['figure.figsize'] = 9, 9
```

Load data from the CSV file

```python
data = dict()
with open('database.csv', 'r') as file_handler:
    csv_data = list(csv.reader(file_handler))
    header = csv_data.pop(0)
    for row in csv_data:
        data[int(row[0])] = {'connections': int(row[1]), 'balance': int(row[2])}
```

# Define the main plot methods

Method to get the minimum and maximun values

```python
def get_scales(data):

    # get all balances and connections
    balances = list()
    connections = list()
    for user_id, user in data.iteritems():
        balances.append(user['balance'])
        connections.append(user['connections'])

    # main vars
    min_balance = np.min(balances)
    max_balance = np.max(balances)
    min_connections = np.min(connections)
    max_connections = np.max(connections)
    
    return min_balance, max_balance, min_connections, max_connections
```

Method to calculate users' position in percentage

```python
def data_to_percent(data):
    
    # get scales
    min_balance, max_balance, min_connections, max_connections = get_scales(data)
    
    # support vars
    balance_scale = max_balance - min_balance
    connections_scale = max_connections - min_connections

    # get position as % (not absolute values)
    balance_percent = list()
    connection_percent = list()
    for user_id, user in data.iteritems():
        balance_percent.append(1 - ((user['balance'] - min_balance) / float(balance_scale)))
        connection_percent.append((user['connections'] - min_connections) / float(connections_scale))
    
    return balance_percent, connection_percent
```

Method to plot

```python
def chart(x, y, title):

    # create a scatter
    fig, ax = plt.subplots()
    ax.scatter(x, y, s=256, color='#20c0f0', alpha=0.2)

    # add titles & labels
    ax.set_title(title, fontsize=13)
    ax.set_xlabel(u'Equilíbrio (%)', fontsize=11)
    ax.set_ylabel(u'Conexões (%)', fontsize=11)

    # add dgrid
    ax.grid(True, linestyle=':', color='0.5')

    # set limits
    ax.set_xlim([-0.05, 1.05])
    ax.set_ylim([-0.05, 1.05])

    # remove axis
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)

    # remove ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
```

# Plot

Plot the main chart:

```python
x, y = data_to_percent(data)
chart(x, y, u'Retrato do Sistêmico (05/02/2015)')
```

Just out of curiosity, let's remove the top 10% more distant from equilibrium

```python
# convert data to a sorted list
sorted_data = sorted(data.iteritems(), key=lambda x: x[1]['balance'])

# filter
users_to_remove = int(round(len(sorted_data) * 0.1)) * -1
sorted_data = sorted_data[:users_to_remove]

#converte data back to a dictionary
data = dict()
for row in sorted_data:
    data[row[0]] = row[1]

# plot
x, y = data_to_percent(data)
chart(x, y, u'Retrato do Sistêmico (05/02/2015)')
```
